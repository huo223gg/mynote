```python
import pandas as pd
import folium
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('market_201809_1.csv', encoding='utf-8')
df2 = pd.read_csv('market_201809_2.csv', encoding='utf-8')
df3 = pd.read_csv('market_201809_3.csv', encoding='utf-8')
df4 = pd.read_csv('market_201809_4.csv', encoding='utf-8')
```

```python
df = pd.concat([df1,df2,df3,df4], ignore_index=True, axis=0)
```

```python
cols=['상호명','위도','경도']
df1 = df[cols]
```

```python
df1
```

```python
df1 = df1[df1['상호명'].str.contains("조마루감자", na=False)]
```

```python
alist=[]
for item in  df1["상호명"]:
    aa = item.split()
    alist.append(aa)
df1["상호명"]=alist
```

```python
df1= df1.reset_index()
```

```python
del df1['index']
```

```python
df1.iloc[0:5,1:2]
```

```python
df1['위도'][1]
```

```python
map = folium.Map(location=[df1["위도"][1], df1["경도"][1]], zoom_start=11)

for n in df1.index:
    folium.CircleMarker([df1["위도"][n], df1["경도"][n]], 
                        radius=1,popup=df1["상호명"][n]).add_to(map)
```
